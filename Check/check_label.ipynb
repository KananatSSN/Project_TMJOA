```python
import os
import pandas as pd
from pathlib import Path

# Path to your folder
folder_path = Path(r"D:\Kananat\Data\Last0\3_Preprocessed")

# Path to your CSV file
csv_file = r"d:\Kananat\Data\Last0\labels.csv"

# Get list of files in the folder and remove the .nii.gz extension
folder_files = set(file.name.replace("_preprocessed.nii.gz", "") for file in folder_path.glob("*.nii.gz"))

# Read file names from CSV
df = pd.read_csv(csv_file)
csv_files = list(df["ID"])

# Find files in folder but not in CSV
in_folder_not_csv = set(folder_files) - set(csv_files)

# Find files in CSV but not in folder
in_csv_not_folder = set(csv_files) - set(folder_files)

# Check for perfect 1-1 correspondence
if len(in_folder_not_csv) == 0 and len(in_csv_not_folder) == 0:
    print("Perfect 1-1 correspondence between folder and CSV file.")
else:
    print("Discrepancies found:")
    if in_folder_not_csv:
        print("Files in folder but not in CSV:", in_folder_not_csv)
    if in_csv_not_folder:
        print("Files in CSV but not in folder:", in_csv_not_folder)

# Print summary
print(f"\nTotal files in folder: {len(folder_files)}")
print(f"Total files in CSV: {len(csv_files)}")
print(f"Files in folder but not in CSV: {len(in_folder_not_csv)}")
print(f"Files in CSV but not in folder: {len(in_csv_not_folder)}")
```
