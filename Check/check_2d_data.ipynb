```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read the image
image = cv2.imread(r"C:\Users\acer\Desktop\Data_2D\train\47-4881 L 2014_z\47-4881 L 2014_z_156.jpg")

# Convert BGR to RGB (OpenCV loads images in BGR format)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Split the image into R, G, B channels
r, g, b = cv2.split(image_rgb)

# Create histograms for each channel, excluding 0
plt.figure(figsize=(15, 5))

for i, color in enumerate(['Red', 'Green', 'Blue']):
    channel = [r, g, b][i]
    
    # Create histogram, excluding 0
    hist = cv2.calcHist([channel], [0], None, [256], [1, 256])
    
    plt.subplot(1, 3, i+1)
    plt.plot(hist, color=color.lower())
    plt.title(f'{color} Channel Histogram')
    plt.xlim([0, 256])
    plt.xlabel('Pixel Intensity')
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
