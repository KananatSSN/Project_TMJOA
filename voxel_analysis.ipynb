```python
import nibabel as nib
import matplotlib.pyplot as plt

def plot_nii_histogram(file_path):
    # Load the .nii file
    nii_image = nib.load(file_path)
    # Get the data from the .nii file
    data = nii_image.get_fdata()
    
    # Flatten the data to 1D for histogram plotting
    data_flat = data.flatten()
    
    # Create a histogram of the pixel values
    plt.figure()
    plt.hist(data_flat, bins=100, color='blue', alpha=0.7)
    plt.title('Histogram of Pixel Values')
    plt.xlabel('Pixel Value')
    plt.ylabel('Frequency')
    
    # Show the plot
    plt.show()

# Example usage

plot_nii_histogram(r"C:\Users\acer\Desktop\Data_Prep_0\imagesTr\5724048_2020_10_01_L.nii")
```
