This script rename the .nii files outputted from 0_dcm_to_nii.py to match the provided Classification.xlsx

```python
nii_folder_path = r"C:\Users\acer\Desktop\TMJOA\Preprocessing\temp0"
label_xlsx_path = r"C:\Users\acer\Desktop\TMJOA\Classification_1.xlsx"
```

```python
import os
import pandas as pd

# Set the path to the folder containing .nii.gz
folder_path = nii_folder_path

# List all files in the directory
file_names = os.listdir(folder_path)

# Create a DataFrame with file names
df = pd.DataFrame(file_names, columns=['nii_file'])

# Edit file name
df['nii_file'] = df['nii_file'].str.split('.').str[0] # remove .nii.gz
df['date'] = df['nii_file'].str.split('_').str[1:4] # extract date 

# Define a function to be use in the apply() method
def rename(filename):
    parts = filename.split('_')
    return f"{parts[0]} {parts[4]}" # return "[PatientID] [L/R]"

# Apply the function to each row
df['nii_file'] = df['nii_file'].apply(lambda x: rename(x))

file_list = df.values.tolist()
print(file_list)
df
```

```python
# Specify the path to the Excel file containing the labels
file_path = label_xlsx_path

# Read the Excel file
excel_df = pd.read_excel(file_path)
excel_df = excel_df[["ID"]]

def extract_date(content):
    parts = content.split(' ')
    return parts[2:] # return ['year', 'month', 'day']

excel_df['Date'] = excel_df['ID'].apply(lambda x: extract_date(x))

def remove_date_from_ID(content):
    parts = content.split(' ')
    return f"{parts[0]} {parts[1]}" # return "[PatientID] [L/R]"

excel_df['ID'] = excel_df['ID'].apply(lambda x: remove_date_from_ID(x))

label_list = excel_df.values.tolist()
print(label_list)
excel_df
```

# Search for .nii.gz corresponding to labels and rename .nii.gz files to match the label

```python
def reconstruct_file_name_from_file_list(file_list_0):
    ID = file_list_0[0].split(' ')[0]
    LR = file_list_0[0].split(' ')[1]
    return f"{ID}_{file_list_0[1][0]}_{file_list_0[1][1]}_{file_list_0[1][2]}_{LR}.nii.gz"
```

```python
def reconstruct_file_name_from_label_list(label_list_0):
    ID = label_list_0[0].split(' ')[0]
    LR = label_list_0[0].split(' ')[1]
    if label_list_0[1] == []:
        return f"{ID} {LR}.nii.gz"
    if label_list_0[1] != []:
        date = []
        for item in range(len(label_list_0[1])):
            date.append(label_list_0[1][item])
        date = ' '.join(date)
    return f"{ID} {LR} {date}.nii.gz"
```

```python
def rename_files_in_folder(folder_path, old_name, new_name):
    print(f"Rename {old_name} to {new_name}")
    old_file = os.path.join(folder_path, old_name)
    new_file = os.path.join(folder_path, new_name)
    os.rename(old_file, new_file)
```

```python
for patient in label_list:
    print(f"Searching for nii files of {patient}")
    patient_Id = patient[0]
    patient_date = patient[1]

    # Search for nii files of this patient
    nii_of_this_patient = []
    for nii in file_list:  
        if nii[0] == patient_Id:
            nii_of_this_patient.append(nii)

    if len(patient_date) == 0 :
        print(f"Found {nii_of_this_patient}")
        if len(nii_of_this_patient) == 1:
            old_name = reconstruct_file_name_from_file_list(nii_of_this_patient[0])
            new_name = reconstruct_file_name_from_label_list(patient)
            rename_files_in_folder(folder_path, old_name, new_name)
        continue

    # Search for nii files of this patient at this year
    nii_of_this_patient_this_year = []
    for nii in nii_of_this_patient:
        nii_date = nii[1]
        if patient_date[0] == nii_date[0]:
            nii_of_this_patient_this_year.append(nii)

    if len(patient_date) == 1 :
        print(f"Found {nii_of_this_patient_this_year}")
        if len(nii_of_this_patient_this_year) == 1:
            old_name = reconstruct_file_name_from_file_list(nii_of_this_patient_this_year[0])
            new_name = reconstruct_file_name_from_label_list(patient)
            rename_files_in_folder(folder_path, old_name, new_name)
        continue

    # Search for nii files of this patient at this year and month
    nii_of_this_patient_this_year_and_month = []
    for nii in nii_of_this_patient_this_year:
        nii_date = nii[1]
        if patient_date[1] == nii_date[1]:
            nii_of_this_patient_this_year_and_month.append(nii)

    if len(patient_date) == 2 :
        print(f"Found {nii_of_this_patient_this_year_and_month}")
        if len(nii_of_this_patient_this_year_and_month) == 1:
            old_name = reconstruct_file_name_from_file_list(nii_of_this_patient_this_year_and_month[0])
            new_name = reconstruct_file_name_from_label_list(patient)
            rename_files_in_folder(folder_path, old_name, new_name)
        continue

    # Search for nii files of this patient at this year and month and day
    nii_of_this_patient_this_year_and_month_and_day = []
    for nii in nii_of_this_patient_this_year_and_month:
        nii_date = nii[1]
        if patient_date[2] == nii_date[2]:
            nii_of_this_patient_this_year_and_month_and_day.append(nii)

    print(f"Found {nii_of_this_patient_this_year_and_month_and_day}")
    if len(nii_of_this_patient_this_year_and_month_and_day) == 1:
        old_name = reconstruct_file_name_from_file_list(nii_of_this_patient_this_year_and_month_and_day[0])
        new_name = reconstruct_file_name_from_label_list(patient)
        rename_files_in_folder(folder_path, old_name, new_name)
    
```

# Remove .nii.gz that don't have label and remove labels that don't have corresponding .nii.gz file

```python
# Specify the path to the Excel file containing the labels
file_path = label_xlsx_path

# Read the Excel file
excel_df = pd.read_excel(file_path)

def remove_if_nii_dont_exist(content):
    file_name = f"{content}.nii.gz"
    file_path = os.path.join(folder_path, file_name)
    print(file_path, os.path.exists(file_path))
    return os.path.exists(file_path) # return True if the file exists

excel_df = excel_df[excel_df['ID'].apply(lambda x: remove_if_nii_dont_exist(x))]
csv_name = "labels_0.csv"
csv_path = os.path.join(folder_path, csv_name)
excel_df.to_csv(csv_path, index=False)
```

```python
df = pd.read_csv(csv_path)
files_to_keep = df["ID"].tolist()

folder_path = nii_folder_path
all_files = os.listdir(folder_path)

for file in all_files:
    file_name = file.split(".")[0]
    if file_name not in files_to_keep:
        file_path = os.path.join(folder_path, file)
        os.remove(file_path)
        print(f"Remove {file_path} because there is no label")
```
