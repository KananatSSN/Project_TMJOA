```python
import pandas as pd

data_dir = r"C:\Users\acer\Downloads\labels.csv"

df = pd.read_csv(data_dir)

df.head()
```

```python
data = df.values.tolist()
ids = [(col[0], col[7]) for col in data]

checked = {}
for id, split in ids:
    raw_id = id.split(' ')[0]
    if raw_id in checked:
        checked[raw_id].add(split)
    else:
        checked[raw_id] = set()
        checked[raw_id].add(split)

ok = all(len(value) == 1 for value in checked.values())
print(ok)
checked
```
