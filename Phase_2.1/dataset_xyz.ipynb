```python
import os
import shutil

def copy_x_folders(source_dir, dest_dir):
    # Ensure the destination directory exists
    os.makedirs(dest_dir, exist_ok=True)

    # Iterate through items in the source directory
    for item in os.listdir(source_dir):
        item_path = os.path.join(source_dir, item)
        
        # Check if it's a directory and has '_x' in its name
        if os.path.isdir(item_path) and '_x' in item:
            # Construct the destination path
            dest_path = os.path.join(dest_dir, item)
            
            # Copy the folder
            shutil.copytree(item_path, dest_path)
            print(f"Copied: {item}")

# Usage
source_directory = r"C:\Users\acer\Desktop\Data_2D\validation"
destination_directory = r"C:\Users\acer\Desktop\Data_2D\data_x\validation"

copy_x_folders(source_directory, destination_directory)
```

```python
import os
import shutil

def move_jpg_images(source_dir, dest_dir):
    # Ensure the destination directory exists
    os.makedirs(dest_dir, exist_ok=True)

    # Walk through all directories and subdirectories
    for root, dirs, files in os.walk(source_dir):
        for file in files:
            # Check if the file is a .jpg image
            if file.lower().endswith('.jpg'):
                # Construct the full file path
                file_path = os.path.join(root, file)
                
                # Construct the destination path
                dest_path = os.path.join(dest_dir, file)
                
                # If a file with the same name already exists in the destination,
                # add a number to the filename
                counter = 1
                while os.path.exists(dest_path):
                    name, ext = os.path.splitext(file)
                    dest_path = os.path.join(dest_dir, f"{name}_{counter}{ext}")
                    counter += 1
                
                # Move the file
                shutil.move(file_path, dest_path)
                print(f"Moved: {file}")

# Usage
source_directory = r"C:\Users\acer\Desktop\Data_2D\data_x\validation"
destination_directory = r"C:\Users\acer\Desktop\Data_2D\data_x\validation_x"

move_jpg_images(source_directory, destination_directory)
```
