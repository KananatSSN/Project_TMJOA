```python
import os
import pandas as pd

# Set the directory path
folder_path = r"C:\Users\acer\Desktop\Data_0"

# List all files in the directory
file_names = os.listdir(folder_path)

# Create a DataFrame with file names
df = pd.DataFrame(file_names, columns=['nii_file'])

# Display the DataFrame
df['nii_file'] = df['nii_file'].str.split('.').str[0]
df['date'] = df['nii_file'].str.split('_').str[1:4]

# Define a function to safely extract elements
def rename(filename):
    # Using a list comprehension to get elements at specified indices if they exist
    parts = filename.split('_')
    return f"{parts[0][:2]}-{parts[0][2:]} {parts[4]}"

# Apply the function to each row in 'ListColumn'
df['nii_file'] = df['nii_file'].apply(lambda x: rename(x))

file_list = df.values.tolist()
print(file_list)
```

```python
def find_duplicates_in_list_of_lists(lists):
    seen = set()
    duplicates = set()
    for sublist in lists:
        tuple_version = tuple(sublist)  # Convert list to tuple to make it hashable
        if tuple_version in seen:
            duplicates.add(tuple_version)
        else:
            seen.add(tuple_version)
    return [list(dup) for dup in duplicates]  # Convert tuples back to lists before returning

# Example usage
list_of_lists = label_list
duplicates = find_duplicates_in_list_of_lists(list_of_lists)
print(duplicates)
```

```python
def reconstruct_file_name_from_file_list(file_list_0):
    ID = file_list_0[0].split(' ')[0]
    LR = file_list_0[0].split(' ')[1]
    return f"{ID[:2]}{ID[3:]}_{file_list_0[1][0]}_{file_list_0[1][1]}_{file_list_0[1][2]}_{LR}.nii.gz"

test = reconstruct_file_name_from_file_list(file_list[0])
print(test)
```

```python
def reconstruct_file_name_from_label_list(label_list_0):
    label_list_0 = [str(item) for item in label_list_0]
    if label_list_0[1] == 'nan':
        label_list_0 = [label_list_0[0]]
    return f"{' '.join(label_list_0)}.nii.gz"

test = reconstruct_file_name_from_label_list(label_list[10])
print(test)
```

```python
def rename_files_in_folder(folder_path, old_name, new_name):
    old_file = os.path.join(folder_path, old_name)
    new_file = os.path.join(folder_path, new_name)
    os.rename(old_file, new_file)
```

```python
import math

for patient in label_list:
    patient_Id = patient[0]

    # Count the number of scans this patient took
    nii_of_this_patient = []
    for nii in file_list:  
        if nii[0] == patient_Id:
            nii_of_this_patient.append(nii)
    
    if len(nii_of_this_patient) == 0 :
        print(f"{patient_Id} does not have any scan")
        print("Debug : missing .nii file **********************************************************")
        print("\n")
        continue

    # Output in case this patient took only 1 scan
    if len(nii_of_this_patient) == 1 and math.isnan(float(str(patient[1]).split(' ')[0])):
        # print(f"{patient_Id} has {len(nii_of_this_patient)} scan")
        # print(nii_of_this_patient)
        print("\n")
        print(nii_of_this_patient[0])
        print(patient)
        old_name = reconstruct_file_name_from_file_list(nii_of_this_patient[0])
        new_name = reconstruct_file_name_from_label_list(patient)
        rename_files_in_folder(folder_path, old_name, new_name)
        print(old_name)
        print(new_name)
        continue
    
    if len(nii_of_this_patient) == 1 :
        scan_year = str(patient[1]).split(' ')[0]
        if int(nii_of_this_patient[0][1][0]) == int(scan_year):
            # print(f"{patient_Id} has {len(nii_of_this_patient)} scan")
            # print(nii_of_this_patient)
            print("\n")
            print(nii_of_this_patient[0])
            print(patient)
            old_name = reconstruct_file_name_from_file_list(nii_of_this_patient[0])
            new_name = reconstruct_file_name_from_label_list(patient)
            rename_files_in_folder(folder_path, old_name, new_name)
            print(old_name)
            print(new_name)
            continue
        elif int(nii_of_this_patient[0][1][0]) != int(scan_year):
            print(f"{patient_Id} does not have any scan from year {scan_year}")
            print(f"This patient {patient}")
            print(f"nii files with this patient ID {nii_of_this_patient}")
            print("Debug : missing .nii file **********************************************************")
            print("\n")
            continue
    
    # Count the number of scans in this year
    scan_year = str(patient[1]).split(' ')[0]
    #print(f"Debug : patient {patient} nii_of_this_patient {nii_of_this_patient} scan_year {scan_year}")
    nii_of_this_patient_this_year = []
    for nii in nii_of_this_patient:
        if nii[1][0] == scan_year:
            nii_of_this_patient_this_year.append(nii)
    
    # Output in case there is only 1 scan in this year
    if len(nii_of_this_patient_this_year) == 1 :
        # print(f"{patient_Id} has {len(nii_of_this_patient_this_year)} scans in year {scan_year}")
        # print(nii_of_this_patient_this_year)
        print("\n")
        print(nii_of_this_patient_this_year[0])
        print(patient)
        old_name = reconstruct_file_name_from_file_list(nii_of_this_patient_this_year[0])
        new_name = reconstruct_file_name_from_label_list(patient)
        rename_files_in_folder(folder_path, old_name, new_name)
        print(old_name)
        print(new_name)
        continue
        
    # Count the number of scans in this month
    scan_month = str(patient[1]).split(' ')[1]
    nii_of_this_patient_this_year_and_month = []
    for nii in nii_of_this_patient_this_year:
        if nii[1][1] == scan_month:
            nii_of_this_patient_this_year_and_month.append(nii)

    # Output in case there is only 1 scan in this month
    if len(nii_of_this_patient_this_year_and_month) == 1:
        # print(f"{patient_Id} has {len(nii_of_this_patient_this_year_and_month)} scans in year {scan_year} and month {scan_month}")
        # print(nii_of_this_patient_this_year_and_month)
        print("\n")
        print(nii_of_this_patient_this_year_and_month[0])
        print(patient)
        old_name = reconstruct_file_name_from_file_list(nii_of_this_patient_this_year_and_month[0])
        new_name = reconstruct_file_name_from_label_list(patient)
        rename_files_in_folder(folder_path, old_name, new_name)
        print(old_name)
        print(new_name)
        continue
    
    # Count the number of scans in this day
    scan_day = str(patient[1]).split(' ')[2]
    nii_of_this_patient_this_year_and_month_and_day = []
    for nii in nii_of_this_patient_this_year_and_month:
        if nii[1][2] == scan_day:
            nii_of_this_patient_this_year_and_month_and_day.append(nii)

    # Output in case there is only 1 scan in this month
    if len(nii_of_this_patient_this_year_and_month_and_day) == 1:
        # print(f"{patient_Id} has {len(nii_of_this_patient_this_year_and_month_and_day)} scans in year {scan_year} and month {scan_month} and day {scan_day}")
        # print(nii_of_this_patient_this_year_and_month_and_day)
        print("\n")
        print(nii_of_this_patient_this_year_and_month_and_day[0])
        print(patient)
        old_name = reconstruct_file_name_from_file_list(nii_of_this_patient_this_year_and_month_and_day[0])
        new_name = reconstruct_file_name_from_label_list(patient)
        rename_files_in_folder(folder_path, old_name, new_name)
        print(old_name)
        print(new_name)
        continue
            
```
