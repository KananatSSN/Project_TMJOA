```python
import timm_3d
import torch
```

```python
model = timm_3d.create_model(
    'resnet18.a1_in1k',
    pretrained=False,
    num_classes=2
)
```

```python
from torchinfo import summary

summary(model,input_size=(1, 3, 224, 224, 224), col_names=["input_size", "output_size", "num_params", "trainable"])  # (batch_size, channels, height, width)
```

```python
print(f"Original first layer: {model.conv1}")
```

```python
import torch.nn as nn

model.conv1 = nn.Conv3d(1, 64, kernel_size=(7, 7, 7), stride=(2, 2, 2), padding=(3, 3, 3), bias=False)
print(f"Changed first layer: {model.conv1}")
```

```python
summary(model,input_size=(1, 1, 224, 224, 224), col_names=["input_size", "output_size", "num_params", "trainable"])  # (batch_size, channels, height, width)
```
